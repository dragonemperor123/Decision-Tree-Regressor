```python
X=df.drop("Chance_of_Admit_",axis=1)
```

```python
y=df["Chance_of_Admit_"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=10)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
```

```python

regressor = DecisionTreeRegressor(random_state=0)
cross_val_score(regressor, X, y, cv=10)
```

```python
regressor.fit(X_train,y_train)
```

```python
regressor.score(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score,r2_score
```

```python
y_scpred=regressor.predict(X_train)
```

```python
print(r2_score(y_train,y_scpred))
```

```python
print(mean_squared_error(y_train,y_scpred))
```

```python
print(mean_absolute_error(y_train,y_scpred))
```

```python
import numpy as np
print(np.sqrt(mean_squared_error(y_train,y_scpred)))
```
